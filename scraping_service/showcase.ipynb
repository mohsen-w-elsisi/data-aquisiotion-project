## Setting up
The project contains multiple scrapers which implement the main `Scraper` interface. Each of the different classes handles a different vendor, however they all implement the `scrape` method which allows for maximal polymorphism and extensability.

Let's work with an example product we would like to shop for: Iphone 17 air.

```python
from dotenv import load_dotenv
from pandas import DataFrame

from scrapers.read_once_list import ReadOnceList

product = "keyboard"

load_dotenv()
```

## Scraping


```python
from constants import SERB_API_KEY
from scrapers import EbayScraper

results = EbayScraper(SERB_API_KEY).scrape(product)

results
```

```python
from scrapers import JumiaScraper

JumiaScraper().scrape(product)
```

```python
from scrapers import NoonScraper
from pandas import DataFrame
from scrapers.read_once_list import ReadOnceList

listings = ReadOnceList()

scraper = NoonScraper()
scraper.push_to(listings)
scraper.scrape(product)

DataFrame(
    [listing.to_json() for listing in listings.read()],
)
```

```python
from composite_scraper import CompositeScraper

listings = CompositeScraper().scrape("toilet")

DataFrame(
    [listing.to_json() for listing in listings],
)
```
